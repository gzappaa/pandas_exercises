### Step 1. Import the necessary libraries

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

chipo = pd.read_csv(url, sep='\t')
```

### Step 3. Assign it to a variable called chipo.

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
```

```python
chipo[chipo['item_price'] > 10.00].shape[0]
```

```python
chipo_copy = chipo.copy()
chipo_copy['product_price'] = chipo_copy['item_price'] / chipo_copy['quantity']
chipo_copy = chipo_copy.drop_duplicates(['item_name','choice_description'])
chipo_copy = chipo_copy.loc[chipo_copy['product_price'] > 10.00, ['item_name','choice_description', 'product_price']].reset_index(drop=True)
chipo_copy
```

### Step 5. What is the price of each item? 
###### print a data frame with only three columns item_name choice_description and product_price

```python
chipo_copy[['item_name','choice_description', 'product_price']]
```

### Step 6. Sort by the name of the item

```python
chipo_copy[['item_name','choice_description', 'product_price']].sort_values('item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.loc[chipo["item_price"].idxmax()]["quantity"]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo.loc[chipo["item_name"] == "Veggie Salad Bowl"]['quantity'].sum()
```
