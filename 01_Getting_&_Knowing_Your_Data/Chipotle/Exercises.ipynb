### Step 1. Import the necessary libraries

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(url, sep='\t')
```

### Step 9. Which was the most-ordered item? 

```python
chipo.groupby('item_name')['quantity'].sum().sort_values(ascending=False).index[0]
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
chipo.groupby('item_name')['quantity'].sum().sort_values(ascending=False).iloc[0]
```

### Step 11. What was the most ordered item in the choice_description column?

```python
res = chipo.groupby('choice_description').sum()
res = res.sort_values(['quantity'], ascending=False).head(1)
res.index[0]
```

### Step 12. How many items were orderd in total?

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo.item_price.dtype
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo.item_price = chipo.item_price.apply(lambda x: float(x[1:]))
```

#### Step 13.c. Check the item price type

```python
chipo.item_price.dtype
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo['quantity'] * chipo['item_price']).sum()
revenue
```

### Step 15. How many orders were made in the period?

```python
orders = chipo.order_id.nunique()
```

### Step 16. What is the average revenue amount per order?

```python
avg = revenue / orders
avg
```
